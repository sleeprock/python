```python
import pandas as pd
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/360340/2_taxi_nyc.csv')
```

```python
taxi = taxi.rename(columns = {'pcp 01':'pcp_01',
                   'pcp 06':'pcp_06',
                   'pcp 24':'pcp_24',
                   })
```

```python
taxi.groupby('borough', as_index=False, dropna=False).pickups.sum().sort_values('pickups')
```

```python
taxi.groupby(['borough', 'hday']) \
     .agg({'pickups': 'mean'})
```

```python
holiday = taxi.query('hday == "Y"').groupby(['borough'], as_index = False).agg({'pickups':'mean'})
workday = taxi.query('hday == "N"').groupby(['borough'], as_index = False).agg({'pickups':'mean'})

for i in range(len(holiday)):
    if holiday.pickups[i] > workday.pickups[i]:
        print(holiday.borough[i])
```

```python
temp = taxi.groupby(['pickup_month', 'borough'], as_index=False).pickups.sum().sort_values(['pickup_month', 'pickups'], ascending=False)
```

```python
def temp_to_celcius(tlist):
    tempc = []
    for i in tlist:
    tempc[i] = ((tlist[i] - 32) * 5) / 9
```

```python
def temp_to_celcius(t):
    t = ((t - 32) * 5) / 9
    return t
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
```
